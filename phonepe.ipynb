```python
import os
import json
import pandas as pd
import psycopg2
import sqlalchemy
from sqlalchemy import create_engine
```

```python
# aggre_insurence

path1="H:/PHPE/Phonepe/pulse/data/aggregated/insurance/country/india/state/"
agg_insurance_list=os.listdir(path1)

colms1={"States":[], "Years":[], "Quarter":[], "Transaction_type":[],"Transaction_count":[],"Transaction_amount":[]}

for state in agg_insurance_list:
    cur_states=path1+state+"/"
    agg_year_list=os.listdir(cur_states)

    for year in agg_year_list:
        cur_year=cur_states+year+"/"
        agg_file_list=os.listdir(cur_year)
        
        for file in agg_file_list:
            cur_file=cur_year+file
            data=open(cur_file,"r")

            A=json.load(data)
            for i in A["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                colms1["Transaction_type"].append(name)
                colms1["Transaction_count"].append(count)
                colms1["Transaction_amount"].append(amount)
                colms1["States"].append(state)
                colms1["Years"].append(year)
                colms1["Quarter"].append(int(file.strip(".json")))


aggre_insurance=pd.DataFrame(colms1)

aggre_insurance["States"]=aggre_insurance["States"].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar') 
aggre_insurance["States"]=aggre_insurance["States"].str.replace("-"," ")
aggre_insurance["States"]=aggre_insurance["States"].str.title()
aggre_insurance["States"]=aggre_insurance["States"].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')

         
```

```python
# Table connection

host="localhost"
user="postgres"
password="Josunny"
database="phonepe"
port="5432"

engine = create_engine(f'postgresql://{user}:{password}@{host}:{port}/{database}')
        
#aggregated_insurance table
aggre_insurance.to_sql(con=engine,name='aggregated_insurance',if_exists='replace',index=False,
                dtype={'States':sqlalchemy.types.VARCHAR(length=225),
                        'Years':sqlalchemy.types.INT,
                        'Quarter':sqlalchemy.types.INT,
                        'Transaction_type':sqlalchemy.types.VARCHAR(length=225),
                        'Transaction_count':sqlalchemy.types.BigInteger,
                        'Transaction_amount':sqlalchemy.types.BigInteger})

#aggregated_transaction table
aggre_transaction.to_sql(con=engine,name='aggregated_transaction',if_exists='replace',index=False,
                dtype={'States':sqlalchemy.types.VARCHAR(length=225),
                        'Years':sqlalchemy.types.INT,
                        'Quarter':sqlalchemy.types.INT,
                        'Transaction_type':sqlalchemy.types.VARCHAR(length=225),
                        'Transaction_count':sqlalchemy.types.BigInteger,
                        'Transaction_amount':sqlalchemy.types.BigInteger})

#aggregated_user table
aggre_user.to_sql(con=engine,name='aggregated_user',if_exists='replace',index=False,
                dtype={'States':sqlalchemy.types.VARCHAR(length=225),
                        'Years':sqlalchemy.types.INT,
                        'Quarter':sqlalchemy.types.INT,
                        'Brands':sqlalchemy.types.VARCHAR(length=225),
                        'Transaction_count':sqlalchemy.types.BigInteger,
                        'Percentage':sqlalchemy.types.FLOAT})

#map_insurance table
map_insurance.to_sql(con=engine,name='map_insurance',if_exists='replace',index=False,
                dtype={'States':sqlalchemy.types.VARCHAR(length=225),
                        'Years':sqlalchemy.types.INT,
                        'Quarter':sqlalchemy.types.INT,
                        'Districts':sqlalchemy.types.VARCHAR(length=225),
                        'Transaction_count':sqlalchemy.types.BigInteger,
                        'Transaction_amount':sqlalchemy.types.BigInteger})

#map_transaction table
map_transaction.to_sql(con=engine,name='map_transaction',if_exists='replace',index=False,
                dtype={'States':sqlalchemy.types.VARCHAR(length=225),
                        'Years':sqlalchemy.types.INT,
                        'Quarter':sqlalchemy.types.INT,
                        'Districts':sqlalchemy.types.VARCHAR(length=225),
                        'Transaction_count':sqlalchemy.types.BigInteger,
                        'Transaction_amount':sqlalchemy.types.BigInteger})

#map_user table
map_user.to_sql(con=engine,name='map_user',if_exists='replace',index=False,
                dtype={'States':sqlalchemy.types.VARCHAR(length=225),
                        'Years':sqlalchemy.types.INT,
                        'Quarter':sqlalchemy.types.INT,
                        'Districts':sqlalchemy.types.VARCHAR(length=225),
                        'Registered_User':sqlalchemy.types.BigInteger,
                        'App_opens':sqlalchemy.types.BigInteger})

#top_insurance table
top_insurance.to_sql(con=engine,name='top_insurance',if_exists='replace',index=False,
                dtype={'States':sqlalchemy.types.VARCHAR(length=225),
                        'Years':sqlalchemy.types.INT,
                        'Quarter':sqlalchemy.types.INT,
                        'Pincodes':sqlalchemy.types.INT,
                        'Transaction_count':sqlalchemy.types.BigInteger,
                        'Transaction_amount':sqlalchemy.types.BigInteger})

#top_transaction table
top_transaction.to_sql(con=engine,name='top_transaction',if_exists='replace',index=False,
                dtype={'States':sqlalchemy.types.VARCHAR(length=225),
                        'Years':sqlalchemy.types.INT,
                        'Quarter':sqlalchemy.types.INT,
                        'Pincodes':sqlalchemy.types.INT,
                        'Transaction_count':sqlalchemy.types.BigInteger,
                        'Transaction_amount':sqlalchemy.types.BigInteger})

#top_user table
top_user.to_sql(con=engine,name='top_user',if_exists='replace',index=False,
                dtype={'States':sqlalchemy.types.VARCHAR(length=225),
                        'Years':sqlalchemy.types.INT,
                        'Quarter':sqlalchemy.types.INT,
                        'Pincodes':sqlalchemy.types.INT,
                        'Registered_Users':sqlalchemy.types.BigInteger})

engine.dispose()
```
